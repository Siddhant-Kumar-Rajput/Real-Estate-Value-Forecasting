# ***Data Visualization***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Cleaned_Data_Pricing.csv')
```

Taking a glance on the data

```python
df.head()
```

Below pairplot give me the idea over the relation between all the numeric value columns in the dataframe

```python
sns.pairplot(df)
plt.show()
```

As I can notice there are some outliers in the data.
Now coming to the relation: First is a slight linear relation between Area and the price, same goes with the BHK and Bathroom. Not much to see in the BHK vs Area and Bathroom vs Area.

```python
sns.pairplot(df,hue='Transaction',palette='Set1')
plt.show()
```

But in the above as we bring Transaction into play, clearly Price increases for newer property as compared to the resale.

```python
sns.pairplot(df,hue='Furnishing',palette='Set1')
plt.show()
```

By performing it with Furnishing, we can see that Semi-Fernished are preferred more by the customers, and price are also more for it.

```python
sns.boxplot(x='BHK',y='Price',data=df, palette='Set1')
plt.show()
```

From the above Box plot, clearly the price of 5 BHK is more as compared to 6 BHK, this is because the area matters. This relation can be seen in pairplot.

```python
sns.lmplot(x='BHK',y='Price',data=df,hue='Transaction',palette='Set1',markers=['*','^'])
plt.show()
```
