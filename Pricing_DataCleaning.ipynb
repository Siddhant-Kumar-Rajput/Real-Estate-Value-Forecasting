Here I checked for the total number of null values present in each column

```python
df.isnull().sum()
```

When the null values are very few as compared to total data, its a best practice to fill those place with the mean value(from the describe()) in case of numeric data only.
For object or bool type data we prefer the mode value.

```python
df['Bathroom']=df['Bathroom'].fillna(df['Bathroom'].mean())
df['Furnishing']=df['Furnishing'].fillna(df['Furnishing'].mode()[0])
```

```python
df['Parking']=df['Parking'].fillna(df['Parking'].mean())
```
